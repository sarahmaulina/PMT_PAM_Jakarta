```python
plt.figure(figsize=(10, 6))
sns.set(style='darkgrid')
sns.distplot(df['pam'], bins=50, color='dodgerblue')
plt.title('Distribusi Penggunaan Air Minum DKI Jakarta 2018-2020', fontsize=20)
plt.axvline(x=df['pam'].median(), linewidth=2, color = 'cornflowerblue')
plt.axvline(x=df['pam'].mean(), linewidth=2, color = 'lightsalmon')
plt.text(931005.0879629629, 0.0000008, 'mean', fontsize=12)
plt.text(621487.0, 0.0000010, 'median', fontsize=12)
plt.xlabel('pam', fontsize=14)
plt.ylabel('frekuensi', fontsize=14)
plt.show()
```

```python
df['pam'].hist(by=df['wilayah'], figsize=(10, 6), bins=20, sharex=True, sharey=True, color='dodgerblue')
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='darkgrid')
sns.distplot(df['pam'], bins=100, color='dodgerblue')
plt.title('Distribusi Penggunaan Air Minum DKI Jakarta 2018-2020', fontsize=20)
plt.xlabel('pam', fontsize=10)
plt.ylabel('frequency', fontsize=10)
plt.show()
```
