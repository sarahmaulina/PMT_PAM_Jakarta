```python
import pandas as pd
from jcopml.utils import save_model, load_model
```

```python
pred = load_model('model/prediction_jakut')
```

```python
kolom = ['tahun', 'bulan', 'wilayah', 'Kecamatan', 'pam', 'pat']
```

```python
data2 = pd.read_csv('datatest_jakut_2.csv')
```

```python
predict = pred.predict(pd.DataFrame(data=data2, columns=kolom))
```

```python
print(predict)
```

```python
prediksi = pd.DataFrame(predict)
```

```python
prediksi.to_csv('prediksi_jakut.csv')
```
